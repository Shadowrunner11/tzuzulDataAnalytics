```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn import datasets

iris = datasets.load_iris()

print(iris.data)
```

```python
print(iris.target)
```

```python
x = iris.data[:,0]

y = iris.data[:,1]
```

```python
species = iris.target

x_min = x.min() - .5
x_max = x.max() + .5

y_min = y.min() - .5
y_max = y.max() + .5

plt.figure()
plt.title('Data set de la flor de Iris, clasificacion por el sepalo')
plt.scatter(x, y, c=species)
plt.xlabel('Largo del sepalo')
plt.ylabel('Ancho del sepalo')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)


plt.show()
```

# DESCOMPOSICION PCA

Principal component analysis

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

x_reducido = PCA(n_components=3).fit_transform(iris.data)

plot_3d = plt.figure()
axes = Axes3D(plot_3d)
axes.scatter(x_reducido[:,0], x_reducido[:,1], x_reducido[:,2], c=iris.target)
axes.set_xlabel('First')
axes.set_ylabel('Second')
axes.set_zlabel('Third')
```

## K-Nearest Neighbors


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

np.random.seed(0)

data = iris.data

target = iris.target

i = np.random.permutation(len(data))

data_train = data[i[:-10]]

target_train = target[i[:-10]]

data_test = data[i[-10:]]
target_test = target[i[-10:]] 

classifier = KNeighborsClassifier()
classifier.fit(data_train, target_train)

print(classifier.predict(data_test))
target_test
```
